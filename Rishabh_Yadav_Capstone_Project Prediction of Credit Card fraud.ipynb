# Handling Imbalanced Dataset

### Undersampling

```python
# Creating a subset of the dataset containing only non-fraudulent transactions
unfraud = data[data['Class']==0]

# Creating a subset of the dataset containing only fraudulent transactions
fraud = data[data['Class']==1]
```

```python
# Getting the dimensions of the subset containing only non-fraudulent transactions
unfraud.shape
```

```python
# Getting the dimensions of the subset containing only fraudulent transactions
fraud.shape
```

```python
unfraud_sample = unfraud.sample(n=473)
```

```python
unfraud_sample.shape
```

```python
# Concatenating the sample subset of non-fraudulent transactions and the subset of fraudulent transactions
new_data = pd.concat([unfraud_sample,fraud],ignore_index=True)
```

```python
new_data['Class'].value_counts()
```

```python
new_data.head()
```

```python
# Creating feature matrix X by dropping the 'Class' column from the concatenated dataset
# Creating target vector y containing only the 'Class' column from the concatenated dataset
X = new_data.drop('Class',axis=1)
y = new_data['Class']
```

```python
# Splitting the concatenated dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,
                                                 random_state=42)
```

# Logistic Regression Model

```python
# Importing Logistic Regression model from sklearn.linear_model
# Creating an instance of Logistic Regression model
# Fitting the model on the training data
from sklearn.linear_model import LogisticRegression
log1 = LogisticRegression()
log1.fit(X_train,y_train)
```

```python
# Generating predictions on the testing data using the trained Logistic Regression model
y_predict1 = log1.predict(X_test)
```

```python
# Importing the accuracy_score function from sklearn.metrics
from sklearn.metrics import accuracy_score
```

```python
# Calculating the accuracy score by comparing the actual target values (y_test) with the predicted values (y_predict1)
accuracy_score(y_test,y_predict1)
```

```python
# Importing precision_score, recall_score, and f1_score functions from sklearn.metrics
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
# Calculating the precision score by comparing the actual target values (y_test) with the predicted values (y_predict1)
precision_score(y_test,y_predict1)
```

```python
# Calculating the recall score by comparing the actual target values (y_test) with the predicted values (y_predict1)
recall_score(y_test,y_predict1)
```

```python
# Calculating the F1 score by comparing the actual target values (y_test) with the predicted values (y_predict1)
f1_score(y_test,y_predict1)
```
